```python
import pandas as pd

import re
```

```python
#ambil file csv yang sudah di generet di twiter
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/sentimen_pcx_nmax.csv")
data.head(10)
```

```python
#menghilangkan tabel gaguna
df = data
df = df.drop('id_str', axis=1)
df = df.drop('quote_count', axis=1)
df = df.drop('reply_count', axis=1)
df = df.drop('retweet_count', axis=1)
df = df.drop('favorite_count', axis=1)
df = df.drop('lang', axis=1)
df = df.drop('user_id_str', axis=1)
df = df.drop('conversation_id_str', axis=1)
df = df.drop('tweet_url', axis=1)
df = df.drop('image_url', axis=1)
df = df.drop('location', axis=1)
df.head()
```

```python
#cleanning dan case folding
import string
def cleaning(kalimat):
  kalimat = re.sub(r'@\w+', '',kalimat)
  kalimat = re.sub(r'https?://\S+', '',kalimat)
  kalimat = re.sub(r'\t|\n|/r', '',kalimat)
  kalimat = kalimat.encode('ascii', 'ignore').decode('ascii')
  kalimat = re.sub(r'@\w/https?://\S+|#\w+', '',kalimat)
  kalimat = re.sub(r'www\.\S+', '', kalimat)
  kalimat = re.sub(r'\d+', '', kalimat)
  kalimat = kalimat.translate(str.maketrans("", "", string.punctuation))
  kalimat = re.sub(r'&quot;', '"', kalimat)
  kalimat = re.sub(r'[^a-zA-Z0-9\s]', '', kalimat)
  kalimat = kalimat.replace('\n', ' ')
  kalimat = kalimat.lower()
  kalimat = ' '.join([word for word in kalimat.split() if len(word) > 1])
  kalimat = ' '.join(re.findall(r'\b\w+\b', kalimat))
  kalimat = kalimat.strip()
  kalimat = re.sub(r'\s+', ' ', kalimat)

  return kalimat
```

```python
df ['text_clean'] = df['full_text'].apply(cleaning)
df.head(10)
```

```python
#text kembar atau duplicate
df = df.drop_duplicates(subset=['text_clean'])
df.head(10)
```

```python
#tokenisasi
import nltk
nltk.download('all')
from nltk.tokenize import word_tokenize
```

```python
def token(kalimat):
  tokens = nltk.tokenize.word_tokenize(kalimat)

  return tokens
```

```python
df['token'] = df['text_clean'].apply(token)
df.head()
```

```python
#stopword
from Sastrawi.StopWordRemover.StopWordRemoverFactory import  StopWordRemoverFactory
factory = StopWordRemoverFactory()
stopwords = factory.get_stop_words()
print(stopwords)
```

```python
def stopwords_text(tokens):
    cleaned_tokens = []
    for token in tokens:
        if token not in stopwords:
            cleaned_tokens.append(token)

    return cleaned_tokens
```

```python
df['stop'] = df['token'].apply(stopwords_text)
df.head()
```

```python
#stemming
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
stem_factory = StemmerFactory()
stemmer = stem_factory.create_stemmer()
```

```python
def stem_text(tokens):
  hasil = [stemmer.stem(token) for token in tokens]
  return hasil
```

```python
df['stemmed'] = df['stop'].apply(stem_text)
df.head()
```

```python
#labeling
lexicon_positive = dict()
import csv
with open('/content/drive/MyDrive/Colab Notebooks/positive.csv', 'r') as csvfile:
  reader = csv.reader(csvfile, delimiter=',')
  for row in reader:
    lexicon_positive[row[0]] = int(row[1])

lexicon_negative = dict()
import csv
with open('/content/drive/MyDrive/Colab Notebooks/negative.csv', 'r') as csvfile:
  reader = csv.reader(csvfile, delimiter=',')
  for row in reader:
    lexicon_negative[row[0]] = int(row[1])

def labeling(kalimat):
    score = 0
    for word in kalimat:
        if word in lexicon_positive:
            score += lexicon_positive[word]
    for word in kalimat:
        if word in lexicon_negative:
            score += lexicon_negative[word]

    polarity = 'positive' if score >= 0 else 'negative'

    return score, polarity
```

```python
df[['polarity_score', 'polarity']] = df['stemmed'].apply(labeling).apply(pd.Series)
# Save the DataFrame to a CSV file
df.to_csv('/content/drive/MyDrive/Colab Notebooks/stemmed.csv', index=False)
df.head()
```

```python
import pandas as pd
import re

# Baca data dari file CSV
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/stemmed.csv')  # Ganti 'data.csv' dengan nama file CSV Anda

# Kolom 'text' berisi dokumen teks
documents = df['stemmed'].tolist()

# Hapus tanda koma, kurung, dan petik dari setiap kata
documents_cleaned = [[' '.join(re.findall(r'\b\w+\b', word)) for word in document.split()] for document in documents]

# Update DataFrame dengan teks yang sudah dibersihkan
df['cleaned_text'] = [' '.join(document) for document in documents_cleaned]

# Tampilkan hasil
print(df['cleaned_text'])
```

```python
import pandas as pd
import math
from sklearn.model_selection import train_test_split

# Assuming df is loaded and contains 'cleaned_text' and 'polarity' columns

# Kolom 'text' berisi dokumen teks
documents = df['cleaned_text'].tolist()

# Bagi data menjadi set pelatihan dan pengujian
data_train, data_test = train_test_split(df[['cleaned_text', 'polarity']], test_size=0.20, shuffle=False)

# Gabungkan kembali data pelatihan ke dalam DataFrame
df_train = pd.DataFrame({'cleaned_text': data_train['cleaned_text'].tolist(), 'polarity': data_train['polarity'].tolist()})

# Fungsi untuk menghitung Term Frequency (TF)
def calculate_tf(term, document):
    term_count = document.count(term)
    total_terms = len(document)
    return term_count / total_terms if total_terms > 0 else 0

# Fungsi untuk menghitung Inverse Document Frequency (IDF)
def calculate_idf(term, document_list):
    document_count = sum(1 for doc in document_list if term in doc)
    total_documents = len(document_list)
    return math.log10(total_documents / (document_count + 1)) if document_count > 0 else 0

# Fungsi untuk menghitung TF-IDF
def calculate_tf_idf(term, document, document_list):
    tf = calculate_tf(term, document)
    idf = calculate_idf(term, document_list)
    return tf * idf

# Daftar term (kata-kata unik) dalam seluruh dokumen pelatihan
all_terms = list(set(term for document in df_train['cleaned_text'] for term in document.split()))

# Membuat DataFrame untuk menyimpan hasil perhitungan TF-IDF
tf_idf_data = []

# Menghitung TF-IDF untuk setiap term dalam setiap dokumen pelatihan
for term in all_terms:
    tf_idf_scores = []
    for document in df_train['cleaned_text']:
        tf_idf_score = calculate_tf_idf(term, document, df_train['cleaned_text'])
        tf_idf_scores.append(tf_idf_score)
    tf_idf_data.append(tf_idf_scores)

# Konversi hasil ke DataFrame
tf_idf_df = pd.DataFrame(tf_idf_data, columns=df_train['cleaned_text'].tolist(), index=all_terms)

# Transpose DataFrame untuk mempermudah pembacaan
tf_idf_df_transposed = tf_idf_df.T

# Reset index for both DataFrames
tf_idf_df_transposed.reset_index(inplace=True)
df_train.reset_index(drop=True, inplace=True)

# Gabungkan dengan kolom 'polarity'
combined_df = pd.concat([tf_idf_df_transposed, df_train['polarity']], axis=1)

# Tampilkan hanya beberapa baris pertama
print(combined_df.head())

# Menampilkan jumlah positif dan negatif
polarity_counts = combined_df['polarity'].value_counts()

# Menampilkan hasil
print("Jumlah Positif:", polarity_counts[1])
print("Jumlah Negatif:", polarity_counts[0])


# Simpan ke file CSV
combined_df.to_csv('/content/drive/MyDrive/Colab Notebooks/tf_idf_results_with_polarity.csv', index=False)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from scipy.special import logsumexp

# Load TF-IDF DataFrame from the saved CSV file
tf_idf_df_transposed = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/tf_idf_results_with_polarity.csv', index_col=0)

# Assuming 'polarity' is the target column
sentiment = tf_idf_df_transposed['polarity']

# Split data into features (X) and labels (y)
features = tf_idf_df_transposed.drop(columns=['polarity'])
labels = tf_idf_df_transposed['polarity']

# Split data into training and testing sets
data_train, data_test, labels_train, labels_test = train_test_split(
    features, labels, test_size=0.20, shuffle=False
)

# Convert features to numpy arrays
train_data = data_train.to_numpy()
test_data = data_test.to_numpy()


class NaiveBayes:
    def __init__(self, classes):
        self.classes = classes
        self.class_priors = {}
        self.cond_probs = {}

    def calculate_prior(self, class_label, train_labels):
        return np.sum(train_labels == class_label) / len(train_labels)

    def calculate_conditional_prob(self, term_index, class_label, train_data, train_labels):
      class_mask = train_labels == class_label
      term_freq = np.sum(train_data[class_mask, term_index] > 0)
      total_docs_in_class = np.sum(class_mask)
      return (term_freq + 1) / (total_docs_in_class + 2)  # Laplace smoothing


    def fit(self, train_data, train_labels):
        for class_label in self.classes:
            self.class_priors[class_label] = self.calculate_prior(class_label, train_labels)
            self.cond_probs[class_label] = {}
            for term_index in range(train_data.shape[1]):
                self.cond_probs[class_label][term_index] = self.calculate_conditional_prob(term_index, class_label, train_data, train_labels)
        print(f'Training - Class: {class_label}, Prior: {self.class_priors[class_label]}')




    def predict(self, test_data):
      predictions = []
      for doc_index, _ in enumerate(test_data):
          doc_log_probs = np.zeros(len(self.classes))
          for class_index, class_label in enumerate(self.classes):
              class_log_prior = np.log(self.class_priors[class_label])
              class_log_probs = [np.log(self.cond_probs[class_label][term_index]) if test_data[doc_index, term_index] > 0 else 0 for term_index in range(test_data.shape[1])]  # Corrected line
              print(f'Doc Index: {doc_index}, Class: {class_label}, Log Prior: {class_log_prior}, Log Probs: {class_log_probs}')
              doc_log_probs[class_index] = class_log_prior + np.sum(class_log_probs)
          predictions.append(self.classes[np.argmax(doc_log_probs)])
      return predictions




# Initialize Naive Bayes classifier
nb_classifier = NaiveBayes(classes=sentiment.unique())

# Train the Naive Bayes classifier
nb_classifier.fit(train_data, labels_train)

# Predict using the trained model
predictions = nb_classifier.predict(test_data)

# Evaluate the model (replace with actual evaluation metrics)
accuracy = np.sum(predictions == labels_test) / len(labels_test)
print(f'Accuracy: {accuracy}')
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Evaluate the model
accuracy = accuracy_score(labels_test, predictions)
precision = precision_score(labels_test, predictions, average='weighted')
recall = recall_score(labels_test, predictions, average='weighted', zero_division=1)
f1 = f1_score(labels_test, predictions, average='weighted')
conf_matrix = confusion_matrix(labels_test, predictions)

# Print the evaluation metrics
print(f'Accuracy: {accuracy}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F1 Score: {f1}')
print('Confusion Matrix:')
print(conf_matrix)
```
